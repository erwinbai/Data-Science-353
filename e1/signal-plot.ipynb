```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N_SAMPLES = 500
```

```python
input_range = np.linspace(0, 2*np.pi, N_SAMPLES, dtype = np. float)
```

```python
signal = np.sin(input_range)
```

```python
noise = np.random.normal(0, 0.7, N_SAMPLES) # random noise with mu=0, sigma=1
```

```python
assert noise.shape == input_range.shape
```

```python
noisy_signal = signal + noise/5
```

```python
plt.plot(input_range, noisy_signal, 'b.')
plt.plot(input_range, signal, 'r-', linewidth = 4)
plt.legend(['Sensor readings', 'Truth'])
plt.xlabel('Time')
plt.ylabel('Value of thing we care about')
plt.show() # of course, your plot will vary slightly, depending on the random noise
```

```python
del signal # With a noisy signal, we dont know the truth: we hope to reconstruct it.
```

```python
from statsmodels.nonparametric.smoothers_lowess import lowess
```

```python
filtered = lowess(noisy_signal, input_range, frac=0.1)
```

```python
plt.plot(input_range, noisy_signal, 'b.')
plt.plot(filtered[:,0], filtered[:,1], 'r-')
plt.legend(['Sensor readings', 'Reconstructed signal'])
plt.show()
```
